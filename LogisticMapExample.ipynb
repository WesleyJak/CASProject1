Example from lecture video on logistic map

```python
import matplotlib.pyplot as plt

# Logistic map Bifurcation diagram
betavals = []
xvals = []

for i in range(0, 401):
    beta = float(i / 100.0)
    xold = 0.5
    xnew = 0
    # transient
    for j in range(0, 2000):
        xnew = beta * xold * (1 - xold)
        xold = xnew
    xss = xnew
    for j in range(0, 1000):
        xnew = beta * xold * (1 - xold)
        xold = xnew
        betavals.append(beta)
        xvals.append(xnew)
        if(abs(xnew-xss))<.001:
            break

plt.scatter(betavals, xvals, s=1)
```

```python
# R value for a stable population. Initial condition x is 0.30
x_old = 0.30
r_val = 2.9
x_vals = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals.append(x_new)
plt.plot(x_vals)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 2.9, x0 = 0.30')
plt.show()
```

```python
# R value for a stable population. Initial condition x is 0.300001
x_old = 0.300001
r_val = 2.9
x_vals = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals.append(x_new)
plt.plot(x_vals)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 2.9, x0 = 0.300001')
plt.show()
```

```python
# R value for a periodic, non-chaotic population. Initial condition x is 0.30
x_old = 0.30
r_val = 3.5
x_vals = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals.append(x_new)
plt.plot(x_vals)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 3.5, x0 = 0.30')
plt.show()
```

```python
# R value for a periodic, non-chaotic population. Initial condition x is 0.300001
x_old = 0.300001
r_val = 3.5
x_vals = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals.append(x_new)
plt.plot(x_vals)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 3.5, x0 = 0.300001')
plt.show()
```

```python
# R value for a chaotic population. Initial condition x is 0.30
x_old = 0.30
r_val = 3.7
x_vals_chaotic1 = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals_chaotic1.append(x_new)
plt.plot(x_vals_chaotic1)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 3.7, x0 = 0.30')
plt.show()
```

```python
# R value for a chaotic population. Initial condition x is 0.300001
x_old = 0.300001
r_val = 3.7
x_vals_chaotic2 = [x_old]
for i in range(0, 200):
  x_new = r_val * x_old * (1 - x_old)
  x_old = x_new
  x_vals_chaotic2.append(x_new)
plt.plot(x_vals_chaotic2)
plt.ylim(0, 1)
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('R = 3.7, x0 = 0.300001')
plt.show()
```

```python
x_vals_chaotic_diff = []

for i in range(1, len(x_vals_chaotic1)):
  x_vals_chaotic_diff.append(x_vals_chaotic1[i] - x_vals_chaotic2[i])


print(x_vals_chaotic1)
print(x_vals_chaotic2)
print(x_vals_chaotic_diff)
```

```python
# a comparison of the .000001 difference in two logistic map plots

plt.plot(x_vals_chaotic1, label='x0 = 0.30')
plt.plot(x_vals_chaotic2, label='x1 = 0.300001')
plt.legend(frameon=True, fontsize=8, loc='upper left')
plt.xlabel('Timesteps')
plt.ylabel('x')
plt.title('Comparison of x0 diff of .000001 at R = 3.7')
plt.show()
```

```python
# difference value between logistic map plots
plt.xlabel('Timesteps')
plt.ylabel('x_n diff')
plt.title('Diff at each timestep between .30 and .300001 for R = 3.7')
plt.plot(x_vals_chaotic_diff)
```
